```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
#To hide Warning messages.
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("googleplaystore.csv")

print("\n First 5 rows in our dataset")
display(df.head(7))
```

```python
df.shape
```

```python
df['Last Updated'] = pd.to_datetime(df['Last Updated'],errors='coerce',format='%B %d, %Y').astype('str')
```

```python
df['Last Updated'].head()
```

```python
def multiple(s):
    try :
        s=list(map(int,s.split('-')))
        return s[0]+(s[1]*12)+s[2]
    except:
        return 'Nan'
```

```python
df['Last Updated']=df['Last Updated'].apply(multiple)
```

```python
df['Last Updated'].head()
```

```python
df['Android Ver']=df['Android Ver'].str.split(n=1,expand=True)
df['Android Ver'].head()
```

```python
df['Android Ver'].isnull().sum()
```

```python
def deal_with_abnormal_strings(s):
    s[s.str.isnumeric()==False]=-1
    s=s.astype(np.float32)
    return s
```

```python
df.Installs=[i.strip().replace('+',"").replace(',','') for i in df.Installs]
df.Installs=deal_with_abnormal_strings(df.Installs)
df.head()
```

```python
df.Size = [i.strip().replace('M', '').replace(',','') for i in df.Size]
df.head()
```

```python
df.Size.unique()
```

```python
def conver_to_float(s):
    try:
        return float(s)
    except ValueError:
        try:
            s=s.split('.')
            return float(s[0]+'.'+s[1])
        except:
            return np.nan
```

```python
df['Size']=df.Size.apply(conver_to_float)
```
